## Homework 2 - Imperfect perturbation

```python
nets=[('P->A->B P->B',{'node_x':[0,1,0.5],'node_y':[0,0,0.86]}),('P->A P->B',{'node_x':[0,1,0.5],'node_y':[0,0,0.86]}),('P->A->B P<-D->B',{'node_x':[0.5,1,0.5,0],'node_y':[0,0,0.86,0],'default_weight':1}),('P->A P<-D->B',{'node_x':[0.5,1,0.5,0],'node_y':[0,0,0.86,0],'default_weight':1.5})]
np.random.seed(123)

display(Markdown('''
Q1. Scenario 2 with $A$->$B$ will show $P$->$A$ and $P$->$B$ but with biased estimation of effect $A$->$B$:
'''))
fig,axes=draw_panel([nets[0]],draw_data=[('P','A'),('P','B')])
plt.show()
```
